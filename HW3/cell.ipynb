```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cell_2 = np.loadtxt('Cell_time_2.txt')
cell_4 = np.loadtxt('Cell_time_4.txt')
cell_8 = np.loadtxt('Cell_time_8.txt')
```

```python
Size_of_array = [4*(i+1) for i in range(50)]
```

```python
def kind_of_speedup(a, b):
    return [b[i]/a for i in range(len(b))]
```

```python
plt.figure(figsize = (8,8))


plt.plot(Size_of_array, kind_of_speedup(2, cell_2), '.-', lw = 2, label = '2 procs')
plt.plot(Size_of_array, kind_of_speedup(4, cell_4), '.-', lw = 2, label = '4 procs')
plt.plot(Size_of_array, kind_of_speedup(8, cell_8), '.-', lw = 2, label = '8 procs')

plt.grid('on')
plt.xlabel('Size', fontsize = 16)
plt.ylabel('Time per proc, s', fontsize = 16)
plt.legend(loc = 'best', fontsize = 14)
plt.show()
plt.savefig('HW3_cell.png')
```

Behaviour is expected - time is increasing with size of array and its trend is decreasing with increasing number of processes. Some outliers are here because the rule and initial conditions are fully random, so sometimes you just get unlucky.
