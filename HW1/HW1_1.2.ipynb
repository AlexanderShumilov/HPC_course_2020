```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def speedup(a, b):
    return [b[i]/a[i] for i in range(len(a))]
```

```python
one = []
busy = []
mutex = []
semaphore = []
n_processes = np.arange(1, 16)
```

```python
i = 0
with open('out.txt') as f:
    for line in f:
        l = line
        time_ind = l.find('time: ')
        if i % 5 == 1: one.append(int(l[time_ind + 6] + l[time_ind + 7]))
        if i % 5 == 2: busy.append(int(l[time_ind + 6] + l[time_ind + 7]))
        if i % 5 == 3: mutex.append(int(l[time_ind + 6] + l[time_ind + 7]))
        if i % 5 == 4: semaphore.append(int(l[time_ind + 6] + l[time_ind + 7]))
        i += 1 
```

```python
plt.figure(figsize = (8,8))


plt.plot(n_processes, speedup(busy, one)[-15:], '.-', lw = 2, label = 'Busy waiting')
plt.plot(n_processes, speedup(mutex, one)[-15:], '.-', lw = 2, label = 'Mutex')
plt.plot(n_processes, speedup(semaphore, one)[-15:], '.-', lw = 2, label = 'Semaphore')

plt.grid('on')
plt.xlabel('# processes', fontsize = 16)
plt.ylabel('Relative speedup', fontsize = 16)
plt.legend(loc = 'best', fontsize = 14)
plt.show()
plt.savefig('HW1.png')
```
